```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
data = pd.read_excel('data.xlsx')
```

**扔掉三列几乎为空的数据**

```python
data = data.drop([u'缩', u'SH上证', u'SH%'], axis=1)
data.drop(2534)
```

**处理日期**

```python
def parseDate(dateStr):
    words = dateStr.strip().split('/')
    return datetime.date(int(words[2]), int(words[0]), int(words[1]))
```

**将数据的日期信息存储起来**

```python
for i in data.index:
    date = parseDate(data.loc[i, u'日期'])
    data.loc[i, 'year'] = date.year
    data.loc[i, 'month'] = date.month
    data.loc[i, 'day'] = date.day
data.sort_values(by=['year', 'month', 'day'], ascending=[1, 1, 1], inplace=True)
data.drop(2534)
```
